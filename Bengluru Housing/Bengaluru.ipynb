```python
"""## Feature Engineering"""
 
# to show all th ecolumns and rows
#pd.set_option("display.max_columns", None)
#pd.set_option("display.max_rows", None)
 
"""### Converting 'total_sqft' cat feature in numeric"""
 
final_dataset['total_sqft'].value_counts()
 
# here we observe that 'total_sqft' contain string value in diff format
#float, int like value 1689.28,817 
# range value: 540 - 740 
# number and string: 142.84Sq. Meter, 117Sq. Yards, 1Grounds
 
# best strategy is to convert it into number by spliting it
 
total_sqft_int = []
for str_val in final_dataset['total_sqft']:
  try:
    total_sqft_int.append(float(str_val)) # if '123.4' like this value in str then conver in float
  except:
    try:
      temp = []
      temp = str_val.split('-')
      total_sqft_int.append((float(temp[0])+float(temp[-1]))/2) # '123 - 534' this str value split and take mean
    except:
      total_sqft_int.append(np.nan) # if value not contain in above format then consider as nan
 
```

```python

final_dataset= final_dataset.reset_index(drop=True) # drop=True - don't add index column in df
# join df4 and total_srft_int list
final_dataset_num = final_dataset.join(pd.DataFrame({'total_sqft_int':total_sqft_int}))
final_dataset_num.head()
```

```python

new_data = final_dataset_num['total_sqft_int'].fillna(final_dataset_num['total_sqft_int'].mean())
```

```python
new_data =final_dataset_num.drop(['total_sqft'],axis =1)
```

```python
new_data.head()
```

```python
new_data['total_sqft_int'] = new_data['total_sqft_int'].fillna(0)

new_data['total_sqft_int'] =new_data['total_sqft_int'].astype(int)
```

```python
X = new_data.drop("price", axis=1)
y = new_data['price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
 #Randomized Search CV

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
# max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
```

```python
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}

print(random_grid)
```

```python
# Use the random grid to search for best hyperparameters
# First create the base model to tune
rf = RandomForestRegressor()
```

```python
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid,scoring='neg_mean_squared_error', 
                               n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)
```

```python
rf_random.fit(X_train,y_train)
```

```python
rf_random.best_params_
```

```python
rf_random.best_score_
```

```python
predictions=rf_random.predict(X_test)
```

```python
sns.distplot(y_test-predictions)
```

```python
plt.scatter(y_test,predictions)
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
